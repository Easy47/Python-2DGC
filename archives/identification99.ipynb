```python
import pandas as pd
import os
import numpy as np
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
PATH = './COVID/'
files = os.listdir(PATH)
```

Molecule list

```python
molecule_list = []

for file in files:
    print(PATH + file)
    df = pd.read_csv(PATH + file)
    tmp = df.loc[df.Formula == 'C6H10O',:]
    for i in range(tmp.shape[0]):
        molecule_list.append(tmp.iloc[i]['Name'])
unique_molecule_list = np.unique(molecule_list)
print(len(unique_molecule_list))
```

Write csv

```python
with open('./res.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    header = files.copy()
    header.insert(0, '')
    writer.writerow(header)
    for molecule in unique_molecule_list:
        res = [molecule]
        for file in files:
            print(PATH + file)
            df = pd.read_csv(PATH + file)
            b = False
            for index, row in df.iterrows():
                if (row['Name'] == molecule):
                    area = (row['Area'])
                    res.append(area)
                    b = True
                    break
            if (not b):
                res.append('')
            
        writer.writerow(res)
```

## (Positif + Positif Faible) vs Negatif

```python
labels = None
files = None
column_max = 11
group_mol_pos_vs_neg = []
for index, row in df.iterrows():

    if (index == 0):
        labels = np.array(row[1:column_max])
        continue
    if (index == 1):
        files = np.array(row[1:column_max])
        continue
    '''g1 = np.array(row[1:11])
    g2 = np.array(row[12:16])'''
    title = row[0]
    row = row[1:column_max]
    g1 = np.array(row[labels == 'negatif'])
    g2 = np.array(row[labels != 'negatif'])

    filtered_g1 = g1[~pd.isnull(g1)]
    filtered_g2 = g2[~pd.isnull(g2)]

    data = [filtered_g1.astype(np.float), filtered_g2.astype(np.float)]
    group_mol_pos_vs_neg.append(data)
    fig = plt.figure(figsize =(5, 4))

    # Creating axes instance
    ax = fig.add_axes([0, 0, 1, 1])
    
    # Creating plot
    bp = ax.boxplot(data)
    plt.title(title)
    # show plot
    plt.show()
```
